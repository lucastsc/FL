```python
import simpy
import random
import itertools
import numpy as np
import matplotlib.pyplot as plt

LINF = 0
LSUP = 10
CLOSE_DISTANCE = 2
class Device:
    def __init__(self, env, nome, numero) -> None:
        self.env = env
        self.nome = nome
        self.numero = numero
        self.x = random.uniform(LINF,LSUP)
        self.y = random.uniform(LINF,LSUP)
        print(f'Device {nome} criado com numero {numero} em ({self.x:.1f},{self.y:.1f})')

    def move(self, devices):
        while True:
            self.x = min(max(self.x + random.uniform(-1, 1), LINF), LSUP)
            self.y = min(max(self.x + random.uniform(-1, 1), LINF), LSUP)
            print(f'Device {self.nome} está em ({self.x:.1f},{self.y:.1f}) com numero {self.numero:.1f} e com vizinhos {[neighbor.nome for neighbor in self.get_neighbors(devices)]}')

            if len(self.get_neighbors(devices)) != 0:
                numeros_vizinhos = [neighbor.numero for neighbor in self.get_neighbors(devices)]
                numeros_totais = numeros_vizinhos + [self.numero]
                media = sum(numeros_totais) / len(numeros_totais)
                self.set_neighbors(devices,media)
                print(f'Devices {[device.nome for device in self.get_neighbors(devices) + [self]]} agora estão com numero {self.numero:.1f} ')

            yield self.env.timeout(1)

    def get_neighbors(self,devices):
        neighbors = []
        for device in devices:
            if ( (((self.x - device.x)**2 + (self.y - device.y)**2) ** 0.5 < CLOSE_DISTANCE) & (self.nome != device.nome) ):
                neighbors.append(device)
        return neighbors
    
    def set_neighbors(self,devices,media):
        neighbors = []
        for device in devices:
            if ( (((self.x - device.x)**2 + (self.y - device.y)**2) ** 0.5 < CLOSE_DISTANCE) & (self.nome != device.nome) ):
                self.numero = media
                device.numero = media

env = simpy.Environment()
devices = [Device(env,f'Device {i}',i) for i in range(10)]
for device in devices:
    env.process(device.move(devices))

env.run(until=100)
```
